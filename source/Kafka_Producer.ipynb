```python
for k in range(len(IdAddress)):
  # Chọn số lượng khách sạn tối đa trên 1 location
  max_loc = 20
  df_test = crawl_link(IdAddress[k][0], IdAddress[k][1], max_loc)
  # Chọn phạm vi của cmt (min_; max_)
  min_cmt = 100
  max_cmt = 500
  df_link = df_test[(df_test['ReviewCount'] > min_cmt) & (df_test['ReviewCount'] < max_cmt)]
  df_link = df_link.set_index(np.array(range(df_link.shape[0])))

  # Tạo dataframe cho từng location
  for i in range(len(df_link)):
    df_temp = df_Hotel(df_link['Address'][i], df_link['HotelName'][i], df_link['HotelCode'][i], df_link['ReviewCount'][i])
    if i == 0: df_location = df_temp
    else: df_location = pd.concat([df_location, df_temp])
  df_location = df_location.set_index(np.array(range(df_location.shape[0])))

  # Nối dataframe của tất cả location
  if k == 0: df_total = df_location
  else: df_total = pd.concat([df_total, df_location])

df_total = df_total.set_index(np.array(range(df_total.shape[0])))
```

```python
df_total.DateStayed = pd.to_datetime(df_total['DateStayed'])
df_total = df_total[df_total['DateStayed'] >= pd.to_datetime('2022-12-20')]
df_total = df_total.sort_values('DateStayed')
df_total.drop('DateStayed', axis=1, inplace=True)
df_total.drop_duplicates(inplace=True)

df_hotel = df_total.set_index(np.array(range(df_total.shape[0])))
```

```python
from kafka import KafkaProducer
from json import dumps
from time import sleep

topic_name = 'TestTopic'
kafka_server = 'localhost:9092'

producer = KafkaProducer(bootstrap_servers=kafka_server,value_serializer = lambda x:dumps(x).encode('utf-8'))

for i in range(len(df_hotel)):
    try:
        data = {'User_name' : df_hotel.User_name[i], 
                'Location_hotel' : df_hotel.Location_hotel[i], 
                'Hotel_name' : df_hotel.Hotel_name[i], 
                'Rating' : df_hotel.Rating[i], 
                'Comment' : df_hotel.Comment[i]}
        producer.send(topic_name, value=data)
        print(" Data mes #" + str(i))
        sleep(5)
        producer.flush()

    except KeyboardInterrupt:
        print("---break---")
        break
print("Live view ended...")
```
